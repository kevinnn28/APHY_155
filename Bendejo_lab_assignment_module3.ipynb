### PROBLEM 1
**Fourier transform of a simple function**

1. Calculate the coefficients in the discrete Fourier transforms of the modulated sine wave $y_n = \sin(\pi n/N) sin(20\pi n/N)$ sampled at $N = 1000$ evenly spaced points using your own user-defined function. *(15 pts.)*

2. Make a plot of their amplitudes similar to the plot shown in Fig. 7.4 of the Newman text. *(15 pts.)*


````python
#Importing necessary code libraries and function
from numpy import zeros, linspace
from cmath import exp, pi, sin
import matplotlib.pyplot as plt

#Calculation for the modulated sinewave for each integer n

#Setting up an array with complex numbers whose elements are initially zeroes
#which will be replaced later by the calculated nth element of y
y=zeros(2000, complex)

#Solving for the y_n
#using for loop to solve for the nth element of y
for n in range (2000): #The range is set to 2000 in order to see the true period of y
  N=1000 #given as the number of evenly spaced points in which we will get our samples
  y[n]=(sin((pi*n)/N))*(sin((20*pi*n)/N)) #assigns the value of theh expression as the nth component of y

#Defining a function that will solve for the Discrete Fourier Coefficients

def dft_coeff(y): #defines the funtion dft_coeff with a parameter y
  N=len(y)        #assigning the length of the array y as N
  c = zeros(N//2+1, complex)    #Setting up an array with complex numbers whose elements are initially zeroes
                                #which will be replaced later by the calculated nth element of y
  #Solving for kth element of the DFT coefficient c
  for k in range (N//2+1):
    for n in range (N):
      #summing the all the nth elements of y that is multiplied to its corresponding complex exponential
      #and assigning it as the kth element of c
      c[k] += (y[n]*exp(-(2j*pi*k*n)/N))
  #this will repeat for all k
  return abs(c) #outputs the absolute values of c


#Plotting the results

fig, ax = plt.subplots(2)
#Plot for y_n
ax[0].plot(y)
ax[0].set_title("Graph of the modulated sine wave")
ax[0].set_xlabel("n")
ax[0].set_ylabel ("y_n")

#Plot for DFT coefficients
ax[1].plot(dft_coeff(y))
ax[1].set_title("Graph of the DFT coefficients")
ax[1].set_xlabel("k")
ax[1].set_ylabel ("c_k")

fig.tight_layout(pad=2.5)
plt.show()


````

````python
#Calculation of discrete fourier transform coefficients using rfft function from numpy"s fft library

#Imports the fft library, and matplotlib.pyplot as plt
from numpy import fft
import matplotlib.pyplot as plt

#Calculating the Discrete Fourier Transform of the data in dow.txt
#and assigning it as coeff_dft
coeff_dft = fft.rfft(dow)

#Verifying that the output (coeff_dft) has the length of (N/2)+1
print ("Length of raw data: " + str(len(dow)), "Length of DFT coefficients: " + str(len(coeff_dft)))

#Setting the last 90% of coeff_dft are zeroes
for n in range (int(len(coeff_dft)*0.1)+1,len(coeff_dft)):
  coeff_dft[n] = 0

#Calculating the Inverse FT Coefficients using the function irrft
#and assigning it as inverse
inverse = fft.irfft(coeff_dft)

#Ploting the raw data and inverse transform of 10% of the DFT Coefficients
fig, ax = plt.subplots()

#plots the Raw data
ax.plot(dow,color='blue', label='Raw Data')

#plots the inverse DFT
ax.plot(inverse, color='orange', label='Inverse FT w/ only 10% DFT coefficients')

#Adding Title and labels for x and y-axis
ax.set_title("Daily closing value of the Dow Jones Industrial Average")
ax.set_xlabel("N-th Day (from late 2006 until the end of 2010)")
ax.set_ylabel ("Closing Value")
ax.legend()
plt.show()
````

The emerging plot of the inverse fourier transform, after setting the last 90% of the Discrete fourier transform coefficients to zeroes, follows the shape of the graph of the original data. However, the graph of the inverse FT has less "noise" or fluctuations than the raw data. It is smoother.

This happened due to the last 90% of the Fourier coefficients being set to 0. The coefficients represent the weight or how much of a certain frequency is in the data. Setting them to zero means that their contributions are being disregarded.
